```python
import numpy as np
import matplotlib.pyplot as plt
```

# Caso $k=1, \Lambda>0$

```python
# Definimos la función f(r)
def f(r, m, k, Lambda):
    return k - 2*m/r - Lambda * r**2 / 3

# Definimos los parámetros
m = 1  # Parámetro m (puede ser ajustado)
r = np.linspace(0.1, 10, 500)  # Rango de r >0 (se evita el punto conflicto r=0).

# Graficamos para los tres casos de Lambda
plt.plot(r, f(r, m,1, 1/(9*m**2)), label=r' $\Lambda=\frac{1}{9m^2}$')
plt.plot(r, f(r, m, 1,0.05), label=r' $\Lambda>\frac{1}{9m^2 }$')
plt.plot(r, f(r, m, 1, 0.5), label=r'$\Lambda<\frac{1}{9m^2} $')
plt.xlabel('r')
plt.ylabel(r'$f(r) = 1-\frac{2m}{r}-\frac{\Lambda r^2}{3}$')
plt.legend()
plt.grid(True)
#plt.xlim(0.1, 5)
plt.ylim(-1, 2)
plt.yticks([0])
plt.xticks([0])
plt.tight_layout()
plt.savefig("ej19grafica1.pdf")
plt.show()
```

# Caso $k=1, \Lambda=0$

```python
plt.plot(r, f(r, m, 1,0), label=r' $m>0$')
plt.plot(r, f(r, 0,1, 0), label=r'$m=0$')
plt.plot(r, f(r, -m,1, 0), label=r'$ m<0$')
plt.xlabel('r')
plt.ylabel(r'$f(r) = 1-\frac{2m}{r}$')
plt.legend()
plt.grid(True)
plt.xlim(0.1, 5)
#plt.ylim(-1, 2)
plt.yticks([0])
plt.xticks([0])
plt.tight_layout()
plt.savefig("ej19grafica2.pdf")
plt.show()
```

```python
plt.plot(r, f(r, m,1, -1), label=r' $m>0$')
plt.plot(r, f(r, 0,1, -1), label=r'$m=0$')
plt.plot(r, f(r, -m,1, -1), label=r'$ m<0$')
plt.legend()
plt.xlabel('r')
plt.ylabel(r'$f(r) = 1-\frac{2m}{r}+\frac{|\Lambda| r^2}{3}$')
plt.grid(True)
#plt.xlim(0.1, 5)
#plt.ylim(-1, 2)
plt.yticks([0])
plt.xticks([0])
plt.tight_layout()
plt.savefig("ej19grafica3.pdf")
plt.show()
```
